# DATASET

```python
data = [12, 15, 14, 10, 18, 20, 17, 16, 14, 13]
```

# TASK 1 – BASIC PYTHON (NO LIBRARIES)

## 1. Print the dataset.

```python
print(data)
```

## 2. Compute manually:
- minimum value
- maximum value
- mean

```python
minimum = data[0]
maximum = data[0]
sum_data = data[0]

for item in data[1:]:
    if item < minimum:
        minimum = item
    if item > maximum:
        maximum = item
    sum_data += item

mean = sum_data / len(data)

print(f"minimum = {minimum}")
print(f"maximum = {maximum}")
print(f"mean = {mean}")
```

## 3. Sort the dataset.

```python
sorted_data = data[:]  # Making a new list object in order to avoid modifying the original data

# Bubble sort
n = len(sorted_data)
for i in range(n):
    for j in range(0, n - i - 1):
        if sorted_data[j] > sorted_data[j + 1]:
            sorted_data[j], sorted_data[j + 1] = sorted_data[j + 1], sorted_data[j]

print(f"data = {data}")
print(f"sorted_data = {sorted_data}")
```

## 4. Compute the median manually.

```python
len_sorted_data = len(sorted_data)

if len_sorted_data % 2 == 0:  # The number of items is even
    median = (sorted_data[len_sorted_data // 2 - 1] + sorted_data[len_sorted_data // 2]) / 2
else:  # # The number of items is odd
    median = sorted_data[len_sorted_data // 2]

print(f"median = {median}")
```

## 5. Find the most frequent value(s) using a dictionary.

```python
count_of_items = {}

for item in data:
    if item not in count_of_items:
        count_of_items[item] = 0
    count_of_items[item] += 1

mode_value = max(count_of_items.items(), key=lambda kv_pair: kv_pair[1])[0]
print(f"mode_value = {mode_value}")
```

# TASK 5 – SMALL FUNCTION

## Write a function called describe_data(data) that returns:
- minimum
- maximum
- mean
- median

```python
import numpy as np

def describe_data(data):
    minimum = np.min(data)
    maximum = np.max(data)
    mean = np.mean(data)
    median = np.median(data)

    print(f"minimum = {minimum}")
    print(f"maximum = {maximum}")
    print(f"mean = {mean}")
    print(f"median = {median}")

describe_data(data)
```
